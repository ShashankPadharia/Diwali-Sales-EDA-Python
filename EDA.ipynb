Null Values Handling 

```python
data["Amount"].fillna(data["Amount"].mean(),inplace=True)
```

```python
data.drop(["Status","unnamed1"], axis=1,inplace=True)
```

```python
data.dropna(inplace=True)
```

```python
data["Amount"]=data["Amount"].astype('int')
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data.rename({"Marital_Status" : "Shadi"},axis=1) #rename column
data.rename(columns={"Marital_Status" : "Shadi"}) #second was to change the column
```

EDA -Exploratiory Data Analysis

Gender

```python
#let get a count of genders
ax = sns.countplot(x="Gender",data=data)



for bar in ax.containers:
    ax.bar_label(bar)       #for adding label in bars
```

Shorting "Gender" values by "Amount" column in descending manner 

```python
data.groupby(['Gender'],as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)
```

```python
sales_gender = data.groupby(['Gender'],as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)
ax = sns.barplot(x="Gender", y= "Amount",data=sales_gender)

for bar in ax.containers:
    ax.bar_label(bar)
```

from above charts we can conclude that most of the buyers are female even purchasing power of females is more than males

##AGE

```python
data.columns
```

Lets check the count of "Genders" by "Age Group"

```python

ax = sns.countplot(x="Age Group" , data=data ,hue="Gender")

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age_group = data.groupby(['Age Group'],as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)

bx = sns.barplot(x="Age Group",y="Amount",data=sales_age_group)

for bar in bx.containers:
    bx.bar_label(bar)
```

From the above chart we can conclude that the 26-35 age group (male-females) range are most buyers

State

Lets Check the total number of orders by Different States

```python
Sales_state = data.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False)


sns.set(rc={"figure.figsize":(25,5)})
cx = sns.barplot(x="State",y="Orders",data=Sales_state)

for bar in cx.containers:
    cx.bar_label(bar)

```

Total Number of Orders by Top 10 States 

```python
Sales_state = data.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)


sns.set(rc={"figure.figsize":(20,5)})  # to change the plot figure size
cx = sns.barplot(x="State",y="Orders",data=Sales_state)

for bar in cx.containers:
    cx.bar_label(bar)    #bar Labels

```

Total amount Spent by Top 10 States

```python
Sales_state = data.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)


sns.set(rc={"figure.figsize":(20,5)})  # to change the plot figure size
cx = sns.barplot(x="State",y="Amount",data=Sales_state)

for bar in cx.containers:
    cx.bar_label(bar)    #bar Labels

```

From Above Charts We can conclude that Sales in Uttar Pradesh,Maharastra,Karnataka states are Higher

Maritial Status

```python
# it Show us order count by Marital Status 


ax = sns.countplot(x="Marital_Status",data=data)

sns.set(rc=({"figure.figsize":(10,5)}))

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#is show Us the Sales by Marital Status and Gender

 
sales_marital = data.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)

ax= sns.barplot(x="Marital_Status",y="Amount",hue="Gender",data=sales_marital)

for bars in ax.containers:
    ax.bar_label(bars)
```

From the Above Graphs We can Conclude that Most buyers are Married (Womens) and they have high purchasing Power

Occupation

```python
sns.set(rc=({"figure.figsize":(20,5)}))

ax = sns.countplot(x="Occupation",data=data)

for bar in ax.containers:
    ax.bar_label(bar)



```

```python
#order Count by occupation

sales_occ=data.groupby(["Occupation"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False)

ax=sns.barplot(x="Occupation",y="Orders",data=sales_occ)

for bar in ax.containers:
    ax.bar_label(bar)
```

From Above Charts we Can conclude that Most orders & buyers are from IT sector,Healthcare,Aviation occupation

```python
#order count by product category

ax=sns.countplot(x="Product_Category",data=data)

sns.set(rc=({"figure.figsize":(25,5)}))

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
#Sales by Product category 

sales_procat=data.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)

ax=sns.barplot(x="Product_Category",y="Amount",data=sales_procat)

sns.set(rc=({"figure.figsize":(25,5)}))

for bar in ax.containers:
    ax.bar_label(bar)
```
